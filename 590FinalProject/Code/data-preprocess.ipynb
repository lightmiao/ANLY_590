```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml' ) 
```

```python
train_dir = 'dfdc_train_part_0'
train_video_files = [x for x in os.listdir(train_dir)]
for i in train_video_files:
    print(i)
    path="output/"+i
    os.mkdir(path)
    video= train_dir +"/"+i 
    count = 0
    cap = cv2.VideoCapture(video)
    ret,frame = cap.read()

    while count < 20:
        cap.set(cv2.CAP_PROP_POS_MSEC,(count*500))   
        ret,frame = cap.read()
       # offset = 50  # play around this value
        face_img = frame.copy()
        
        face_rects = face_cascade.detectMultiScale(face_img,scaleFactor=1.3, minNeighbors=5) 

        for (x,y,w,h) in face_rects: 
            roi = face_img[y:y+h,x:x+w] 
        
        cv2.imwrite(path+"/frame%d.jpg" % count, roi)     # Next I will save frame as JPEG
        count = count + 1
```
